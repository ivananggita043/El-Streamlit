```python
import pandas as pd
```

```python
obesitas = pd.read_csv('obesitasclean.csv')
```

```python
data = obesitas.copy()
```

```python
data
```

```python
target = 'Label'
```

```python
#kolom gender akan di encode menjadi binary true atau false
encode = ['Gender']
```

```python
for col in encode:
    dummy = pd.get_dummies(data[col], prefix=col)
    data = pd.concat([data,dummy], axis=1)
    del data[col]
```

```python
data
```

```python
#selanjutntya memetakan label target menjadi numerik
target_mapper = {'Underweight':0, 'Normal Weight':1, 'Overweight':2, 'Obese':3}
```

```python
def target_encode(val):
    return target_mapper[val]
```

```python
data['Label'] = data['Label'].apply(target_encode)
```

```python
data
```

```python
# split data menjadi 2, X dan Y dimana
# X merupakan seluruh dataset tanpa kolom label dan
# Y merupakan data jenis saja
X = data.drop('Label', axis=1)
Y = data['Label']
```

```python
X
```

```python
Y
```

```python
#Memodelkan dataset menggunakan NBC
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X, Y)
```

```python
#Menyimpan Model NBC
import pickle
pickle.dump(model,open('modelNBC_obesitas.pkl','wb'))
```
